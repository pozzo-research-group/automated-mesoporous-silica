```python
import os
import sys
sys.path.append('../sample_utilities/')
sys.path.append('..')
#import stober_synthesis_utils
import matplotlib.pyplot as plt
import pandas as pd
from sample_utilities import samples
import uuid
import json
```

```python
target_volume = 1700
n_samples = 40
target_dilution = 0.048 # mol Si/L saxs sample
constants_fp = './constants_2024_10_14_highlights_replicates.json'
samples_write_fp = 'Sample_table_2024_10_14_highlights_replicates.csv'
previous_sampeles_fp = ['../2024_10_04_RandomBaseline_v2/Sample_table_2024_10_04_randomsampleV2.csv', '../2024_10_07_randomSampleV3/Sample_table_2024_10_07_randomsampleV3.csv']
```

```python
## load selected sample table with uuids

selected_samples = pd.read_csv('selected_samples.md', sep= '|')
```

```python
selected_uuids = set(selected_samples[' UUID                                 '].iloc[1:])
```

```python
selected_uuids = set(item.strip() for item in selected_uuids)
```

```python
selected_uuids
```

```python
## loop over previous samples to get parameters for uuid
existing_compositions = set()
previous_sampletables = []
for fp in previous_sampeles_fp:
    df = pd.read_csv(fp)
    previous_sampletables.append(df)

    for i, row in df.iterrows():
        sample = (int(row.teos_volume), int(row.ammonia_volume), int(row.water_volume))
        existing_compositions.add(sample)

```

```python
selected_sample_rows = []

previous_samples = pd.concat(previous_sampletables, axis = 0)

#for sample_uuid in selected_uuids:
    

selected_sampletable = previous_samples[previous_samples['uuid'].isin(selected_uuids)]
```

```python
selected_sampletable
```

```python
selected_sampletable = selected_sampletable.drop('well', axis = 1)
```

```python
## generate new UUIDs for these samples and keep mapping

new_uuids = []
uuid_mapper = {}
for val in selected_sampletable.uuid:
    new_uuid = uuid.uuid4()
    uuid_mapper[val] = str(new_uuid)
    new_uuids.append(str(new_uuid))
    
```

```python
selected_sampletable['uuid'] = new_uuids
```

```python
with open('replicate_sample_uuid_mapper.json', 'wt') as f:
    json.dump(uuid_mapper, f)
```

```python
uuid_mapper
```

```python
selected_sampletable
```

```python
## generate new random sample to get to 40:
samples_gen = []
while len(samples_gen) < n_samples - len(selected_sampletable):
    sample = samples.SolidSilicaSample(target_volume, reactant_fp=constants_fp)
    sample.generate_random_vol_fractions()
    sample.calculate_reactant_volumes()

    #check for dupes
    sample_comp = (int(sample.teos_volume), int(sample.ammonia_volume), int(sample.water_volume))

    if sample_comp in existing_compositions:
        print(f'found existing composition: {sample_comp}')
        continue
    else:
        existing_compositions.add(sample_comp)
        sample.calculate_dilution_volumefraction(target_dilution)
        sample.calculate_silica_mass_concentration()
        sample.calculate_silica_mass_fraction()
        samples_gen.append(sample)
```

```python
compiled_samples = pd.concat([selected_sampletable, samples.generate_synthesis_table(samples_gen)], axis = 0)
```

```python
compiled_samples
```

```python
# round b/c we can't dose fractional uL

cols = ['teos_volume', 'ammonia_volume', 'water_volume', 'ethanol_volume']
compiled_samples[cols] = compiled_samples[cols].round()
```

```python
compiled_samples
```

```python
#assign rows to sample table
rows = ['A', 'B', 'C', 'D', 'E', 'F']
cols = [1, 2, 3, 4, 5, 6, 7, 8]

wells = []

#skip [B,E]x[2,7]
for row in rows:
    for col in cols:
        if row in ['B', 'E'] and col in [2, 7]:
            pass
        else:
            wells.append(row + str(col))
```

```python
compiled_samples['well'] = wells[0:len(compiled_samples)]
```
