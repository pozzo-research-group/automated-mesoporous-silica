```python
current_fp = "/mnt/c/Users/bgpel/My Drive/Research/PozzoResearchGroup/APS_Beamtime_Sp25/BatchSamplesData/Reduced_ascii_4_18_25_emptyBlank"
new_fp = "/mnt/c/Users/bgpel/My Drive/Research/PozzoResearchGroup/APS_Beamtime_Sp25/BatchSamplesData/Reduced_ascii_4_18_25_emptyBlank_renamed"
```

```python
with open(current_fp + '/A3_1_2036.dat', 'rt') as f:
    for line in f:
        print(line)
```

```python
for i, row in sample_table.iterrows():
    if row['Loaded?'] == 'y' and row['Sample ID'] is not np.nan:
        cell = row['Cartridge cell']
        plate = int(row['Cartridge plate'])
        uid = row['Sample ID']
        if plate == 1:
            platename = '0'
        elif plate == 2:
            platename = '0_c2'
        fp_stub = f'{cell}_{platename}*'
        print(fp_stub)
        files = glob.glob(os.path.join(current_fp, fp_stub))
        if len(files) !=1:
            #raise AssertionError(f'Error on file {fp_stub}, returned files {files}')
            if 'c2' in fp_stub:
                files = [f for f in files if 'c2' in f]
            else:
                files = [f for f in files if 'c2' not in f]
        print(files)
        old_name = files[0].split('/')[-1].split('.')[0]
        if i < 10:
            i = f'0{i}'
        new_name = f'{i}_{uid}_{old_name}.dat'
        print(new_name)

        shutil.copy(files[0], new_fp + '/' + new_name)
```
